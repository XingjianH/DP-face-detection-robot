#### Fonctions

Exemple de fonction : permutation des 2 variables qui lui sont passées en argument.

```python
def permutation(a, b):
   tmp = a
   a = b
   b = tmp
   return a, b

x, y = 1, 2
x, y = permutation(x, y)
print(x, y)
```

On constate que la fonction a bien fait son opération de permutation.
On se souvient qu'avec le langage C, on avait besoin de faire un passage par adresse pour faire ça.

On remarque qu'il n'y a ni début ni fin à la fonction.  
Ce sont les identations qui font office de début.   
Il en sera de même pour les boucles, les tests, etc.

## Classes

Exemple :

```python
import math

class nombre_complexe:
    def __init__(self, a=0, b=0):
        self.a, self.b = a, b
    def get_module(self):
        return math.sqrt(self.a * self.a + self.b * self.b)
    def ajoute(self, c):
        return nombre_complexe(self.a + c.a, self.b + c.b)

c1 = nombre_complexe(0, 1)
c2 = nombre_complexe(1, 0)

print(c1.get_module())
c = c1.ajoute(c2)         # c = c1 + c2
print(c.a, c.b)
```

#### Image en niveaux de gris

Pour avoir une image en niveaux de gris, il suffit de donner les mêmes valeurs aux 3 teintes R, G, B.

Exemple : dégradé du noir au blanc de la gauche vers la droite :

```python
array = np.zeros([256, 256, 3], dtype=np.uint8)
for x in range(256):
    for y in range(256):
        array[y, x, 0] = array[y, x, 1] = array[y, x, 2] = x
img = Image.fromarray(array)
plt.imshow(img)
```

#### 2) Analyse en Composantes principales (ACP)

Cette méthode consiste à projeter les donnnées sur un espace de même dimensions que les leurs, mais en maximisant la variance sur les premières dimensions. En d'autres termes, on peut dire que l'information est ramenée au maximum dans ces premières dimensions (= les "composantes principales"). On peut alors ne conserver que ces premières dimensions, par exemple pour l'affichage, qui reflètent le mieux possible le contenu des données.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import decomposition
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(5)
centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

fig = plt.figure(1, figsize=(8, 6))
#ax = Axes3D(fig, rect=[0, 0, .95, 1])
ax = plt.axes(projection="3d", elev=48, azim=134)
for name, label in [('Setosa', 0), ('Versicolor', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(), X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,  horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor='k')
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])
plt.show()
```

#### 3) Méthode de Sammon

Comme pour l'ACP, la méthode de Sammon projette les données dans un sous-espace (dont les dimensions sont définies à l'avance, par exemple 2 ou 3), en essayant de minimiser l'écart entre les distances mesurées dans l'espace de départ et celles mesurées dans l'espace d'arrivée. Cette minimisation est obtenue au moyen d'un algorithme itératif basé sur la descente de gradient, appliquée à ce critère. Les distances dont il est question concernent les vecteurs de donnée pris 2 à 2.

L'implémentation Python utilisée ici est celle qui est disponible au lien suivant : https://github.com/tompollard/sammon

Rm : 
- la fonction "sammon" est définie dans le fichier sammon.py, qui doit se trouver dans le même répertoire que ce notebook
- plus généralement, pour inclure une fonction "ma_fonction" présente dans le fichier "mon_fichier", lui-même présent dans le répertoire "mon_rep", on écrit : 

from mon_rep.mon_fichier import ma_fonction
- les distances entre les points dans l'espace d'arrivée sont initialisées aléatoirement ; chaque exécution de l'algorithme donnera donc un résultat différent.

```python
!wget https://raw.githubusercontent.com/tompollard/sammon/master/sammon.py
```

```python
from sammon import sammon        #import the sammon frunction from sammon.py
import numpy as np 
from sklearn import datasets     #to use the iris dataset
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

"""
   Description : This notebook tests sammon.py by applying it
   				  to Fisher's iris dataset
                 http://en.wikipedia.org/wiki/Iris_flower_data_set
"""
# Load the iris data
iris = datasets.load_iris()
(x,index) = np.unique(iris.data,axis=0,return_index=True)
target = iris.target[index]
names = iris.target_names

# Run the Sammon projection
[y,E] = sammon(x, 2, init='random')

# 2D Plot
plt.scatter(y[target ==0, 0], y[target ==0, 1], s=20, c='r', marker='o',label=names[0])
plt.scatter(y[target ==1, 0], y[target ==1, 1], s=20, c='b', marker='D',label=names[1])
plt.scatter(y[target ==2, 0], y[target ==2, 1], s=20, c='y', marker='v',label=names[2])
plt.title('Sammon projection of iris flower data')
plt.legend(loc=2)
plt.show()
```

Interprétation de ce résultat : 
- on retrouve ce qu'on a déjà observé avec les 2 autres types d'affichage, à savoir que la classe "setosa" est bien distincte des deux autres, et donc sera plus facile à classifier (quand on fera de la classification)

Pour traiter toute la base d'images avec des algorithmes de classification, if faudra charger toute la base en mémoire.

Tant qu'on n'utilise pas de réseau de neurones convolutif, on se contentera de charger les images en niveaux de gris. On devra donc convertir leur format, car elles sont en couleur.

La taille de ces images étant variable, il faut également les ramener à une même taille (dont le choix est libre).

On peut ajouter des fonctions timer pour mesurer et afficher le temps de chargement.

```python
import os
import glob
import cv2
import numpy as np

data_path = 'data/GTSRB/Final_Training/Images/'

nb_classes = 43
rows, cols = 48, 48       #nouvelle taille des images

def load_GTSRB():
    print("Load data...")
    data = []
    labels = []
    for i in range(nb_classes):
        image_path = data_path + '/' + format(i, '05d') + '/'
        print("chargement répertoire", image_path)
        for img in glob.glob(image_path + '*.ppm'):
            image = cv2.imread(img)
            #print(img)
            image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
            image = (image / 255.0)                            #rescale
            image = cv2.resize(image, (rows, cols))            #resize
            image = np.array(image, 'f')
            image = image.reshape(rows*cols)                   #data flattening (aplatissement) 
            data.append(image)
            label = int(image_path.split('/')[-2])
            labels.append(label)
    print('OK')
    return (data, labels)

X, y = load_GTSRB()
print(np.shape(X))
print(np.shape(y))
print(type(X))
print(type(y))
```

### Egalisation d'histogramme (Exercice 1.3)

Les images GTSRB sont des images prises en conditions réelles, pour lesquelles l'égalisation histogremme, qui permet de  normaliser le contraste, peut être pertinent, selon le classifieur utilisé.
On peut utiliser pour ça la fonction "equalize_hist" de Scikit-Image.

Attention : les répertoires des images contiennent tous un fichier CSV. C'est donc à prendre en compte au moment du chargement de la base complète (cf source ci-dessous). Pour afficher quelques images, ça n'est pas gênant.

Sources utiles : 
https://chsasank.github.io/keras-tutorial.html

Exemple de résultats obtenu, si on choisit d'afficher la 1ère image de chaque sous-répertoire :

<td> <img src="GTSRB.png" width =500 height=500> </td>
<td> <img src="GTSRB_egal_histo.png" width =500 height=500> </td>
</tr></table>

```python
import os
import glob
import cv2
import numpy as np
from skimage import color, exposure, transform
from skimage import io
import matplotlib.pyplot as plt

mon_rep = 'data/GTSRB/Final_Training/Images/'

liste_rep = sorted(os.listdir(mon_rep))
print(liste_rep)
for rep in liste_rep:
    fichiers = sorted(os.listdir(mon_rep + rep))

mon_fichier = mon_rep + liste_rep[2] + '/' + fichiers[1]
print(mon_fichier)
img = io.imread(mon_fichier)

 # Histogram normalization in v channel
hsv = color.rgb2hsv(img)
hsv[:, :, 2] = exposure.equalize_hist(hsv[:, :, 2])
img = color.hsv2rgb(hsv)

plt.imshow(img)
```
